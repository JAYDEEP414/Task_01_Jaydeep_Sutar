# Load Dataset

```python
import pandas as pd

df = pd.read_excel("Dataset.xlsx")
print(df.head())
print(df.info())
```

# Checking Missing Values

```python
print(df.isnull().sum())
```

# Handling Missing Coupon Codes

```python
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

```python
print(df.isnull().sum())
```

# Cheaking Duplicate Rows

```python
print("Duplicate rows:", df.duplicated().sum())
```

# Verifying Unique Order IDs

```python
print("Duplicate OrderIDs:", df["OrderID"].duplicated().sum())
```

# Standardizing Text Columns

```python
text_cols = [
    "Product", "PaymentMethod", "OrderStatus",
    "CouponCode", "ReferralSource"
]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

# Format Date Column

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Date"] = df["Date"].dt.strftime("%Y-%m-%d")
```

# Validate TotalPrice

```python
df["CalculatedTotal"] = df["Quantity"] * df["UnitPrice"]
df["PriceMatch"] = df["CalculatedTotal"].round(2) == df["TotalPrice"].round(2)

print("Incorrect TotalPrice rows:", (~df["PriceMatch"]).sum())
```

# Removing Temporary Columns

```python
df.drop(columns=["CalculatedTotal", "PriceMatch"], inplace=True)
```

# Final Validation

```python
print("Missing values:", df.isnull().sum().sum())
print("Duplicate rows:", df.duplicated().sum())
print("Duplicate OrderIDs:", df["OrderID"].duplicated().sum())
```

# Saving Cleaned Dataset

```python
df.to_csv("Cleaned_Dataset.csv", index=False)
```
